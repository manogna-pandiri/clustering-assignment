THEORETICAL QUESTIONS

1. What is unsupervised learning in the context of machine learning?
ANSWER :

Unsupervised learning is a branch of machine learning in which the model discovers patterns in unlabeled data, or data without output labels.

Finding hidden relationships, patterns, or structures in the data is the aim.

Typical tasks:
- Clustering is the process of putting related data points together (e.g., K-Means, DBSCAN).
- Reducing the number of features without sacrificing information is known as dimensionality reduction (e.g., PCA, t-SNE).

Example, consider customer segmentation in marketing, which groups clients according to their purchasing patterns without revealing categories in advance.


2. How does K-Means clustering algorithm work?
ANSWER :

K-Means is a partition-based clustering algorithm that divides data into K clusters.

STEPS :
1. Choose the number of clusters K.
2. Initialize K centroids (cluster centers) at random.
3. Each data point should be assigned to the closest centroid.
4. Determine the mean of all the points allocated to that cluster in order to update the centroids.
5. Continue steps 3–4 until centroids stabilize, meaning they change little or not at all.

Key Points:
- Spherical clusters are the most effective.
- Sensitive to the location of the centroid at first.


3. Explain the concept of a dendrogram in hierarchical clustering.
ANSWER :

A dendrogram is a diagram that resembles a tree and illustrates the hierarchical connections between clusters or data points.

- A single data point is represented by each leaf.
- Branches, or nodes, indicate whether a cluster is dividing (divisive) or merging (agglomerative).
- The distance or difference between clusters is represented by the height of a branch.
- To obtain the desired number of clusters, you can "cut" the dendrogram at a specific height.

Use: In hierarchical clustering, it aids in determining the number of clusters and visualizing their structure.


4. What is the main difference between K-Means and Hierarchical Clustering?
ANSWER :

The difference between Hierarchical Clustering and K-Means is in the way they cluster data. K-Means separates data into a fixed number of clusters you decide before running it. It operates by mapping points to cluster centroids and shifting those centroids until they converge. It's quick and performs well when clusters are relatively round and equal in size. Hierarchical Clustering, however, constructs a tree-like structure known as a dendrogram that indicates how data points cluster at varying levels of hierarchy. You have the option of determining how many clusters you desire by pruning the tree at a specific height. It can deal with clusters that are not of regular shape but is slower when dealing with big data sets. Thus, K-Means provides a simple set of clusters, whereas Hierarchical Clustering reveals an entire hierarchy of relations in the data.


5.What are the advantages of DBSCAN over K-Means?
ANSWER :

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is better than K-Means in the following ways:

- It can discover clusters of any shape, not only spherical clusters.
- It identifies noise or outliers automatically, rather than assigning all points to a cluster.
- You don't have to pre-specify the number of clusters.
- Performs well when clusters differ in size and density.


6. When would you use Silhouette Score in clustering?
ANSWER :

The Silhouette Score estimates how well a data point belongs to its designated cluster relative to other clusters.

- You apply it to measure the clustering quality.
- It's useful for Determining the ideal number of clusters and comparing clustering algorithms.
- Score between -1 and 1, with higher values meaning better clusters.


7. What are the limitations of Hierarchical Clustering?
ANSWER :

- Computationally intensive with large datasets.
- Sensitive to outliers and noise, which can alter the hierarchy.
- Once a merge or split, it can never be reversed (no backtracking).
- Selecting the appropriate number of clusters may be subjective, often by way of dendrogram interpretation.


8. Why is feature scaling important in clustering algorithms like K-Means?
ANSWER :

Clustering algorithms such as K-Means are based on distance measures (typically Euclidean distance).

- If features are at different scales, features that have larger ranges will overbear the computation of distance.

- Scaling (i.e., standardization or normalization) is used to make all features contribute equally towards more significant clusters.


9. How does DBSCAN identify noise points?
ANSWER :

DBSCAN detects points as noise when they are not satisfying the density requirement:

- A point is a core point if it contains at least minPts points in its EPS radius.
- Points that are not core points and are not in EPS of any core point are noise.


10. Define inertia in the context of K-Means.
ANSWER :

Inertia is a measure of compactness of clusters for K-Means.

- It is the total squared distance between each point and the centroid of the corresponding cluster.
- Lower inertia indicates points are closer to their centroids, meaning closer clusters.
- It is commonly utilised to assist in determining the number of clusters using the elbow method.


11. What is the elbow approach in K-Means clustering?
ANSWER :

The elbow approach is a method for selecting the best number of clusters K.
K in K-Means.
- You apply K-Means with varying values of K and compute inertia (sum of squared distances).
- Plot K vs. inertia.
- The place where the rate of decrease in inertia drops noticeably, creating an "elbow" on the plot, is thought to be the optimal number of clusters.


12. Describe the concept of "density" in DBSCAN.
ANSWER :

Density in DBSCAN is the number of points in a given radius (EPS) from a data point.
- A dense area contains sufficient points (≥ minPts) in EPS.
- Dense areas constitute clusters, while points in low-density areas are treated as noise.

13. Can hierarchical clustering be used on categorical data?
ANSWER :

Yes, but you'll need to use a distance measure appropriate for categorical data (such as Hamming distance or Jaccard similarity) rather than Euclidean distance.

- Hierarchical clustering will work on any data provided that you can define a useful similarity/distance measure.
- Standard implementations tend to assume numeric data, so you may need to do some preprocessing.


14. What does a negative Silhouette Score indicate?
ANSWER :

A negative Silhouette Score indicates that a point is nearer to points of a different cluster than its own cluster.
- This signifies poor clustering or points assigned in error.
- Silhouette scores between -1 and 1:
  Close to 1 → well-clustered
  Close to 0 → on the boundary
  Negative → probably in the wrong cluster


15. Explain the term "linkage criteria" in hierarchical clustering.
ANSWER :

Linkage criteria decide how the distance between clusters is measured in hierarchical clustering.
Common types:
- Single linkage: Distance between two clusters' closest points.
- Complete linkage: Distance between two clusters' farthest points.
- Average linkage: Average of the distance for all pairs of points in two clusters.
- Ward's linkage: Reduces the variance within clusters while joining.


16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
ANSWER :

K-Means makes the assumption that clusters are approximately spherical and of comparable sizes. If clusters are of different sizes or densities:

- K-Means might divide large clusters or combine small ones in the wrong way.
- It is based on distance to the centroids, and thus points in sparse areas may be mistakenly assigned to close-by dense clusters.


17. What are the core parameters in DBSCAN, and how do they influence clustering?
ANSWER :

 1.DBSCAN has two primary parameters:
- EPS (epsilon): The neighborhood radius around a point to count its neighbors.
- Greater EPS → more points are considered → fewer, larger clusters.
- Small EPS → tight clusters, more points are considered as noise.

2. minPts: The minimum number of points needed to create a dense area.
- Greater minPts → only extremely dense areas will form a cluster → more noise.
- Smaller minPts → more points are included → potential to merge sparse areas.

EPS and minPts work together to constrain the shape, size, and number of clusters, and which points are noise.


18. How does K-Means++ improve upon standard K-Means initialization?
ANSWER :

K-Means++ enhances the initial choice of centroids to minimize the likelihood of poor clustering.
- Rather than selecting centroids randomly, it distributes them evenly, assigning greater probability to points distant from current centroids.
- This results in quicker convergence and sometimes more optimal final clusters.


19. What is agglomerative clustering?
ANSWER :

Agglomerative clustering is hierarchical clustering that proceeds bottom-up:
- Every point begins as an individual cluster.
- In every step, the two nearest clusters are combined based on a linkage rule.
- This goes on until all the points are part of one cluster or until a stopping criterion (such as number of desired clusters) is reached.


20. What makes Silhouette Score a better metric than just inertia for model evaluation?
ANSWER :

- Inertia only considers the proximity of points to their cluster centers, without concern for how separate clusters are from each other.

- Silhouette Score takes into account both cohesion (distance within a cluster) and separation (distance to the nearest other cluster).

- This provides a more even assessment of clustering quality, particularly where clusters are dissimilarly sized or shaped.

```python
# 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

# Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Visualize the clusters
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title('K-Means Clustering on Synthetic Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# 23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot.
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np

# Generate synthetic moon-shaped data
X, y_true = make_moons(n_samples=500, noise=0.05, random_state=42)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
y_db = dbscan.fit_predict(X_scaled)

# Plot clusters and highlight outliers
plt.figure(figsize=(8,6))

# Mask for noise points (label = -1)
noise_mask = y_db == -1
cluster_mask = y_db != -1

# Plot clusters
plt.scatter(X_scaled[cluster_mask, 0], X_scaled[cluster_mask, 1],
            c=y_db[cluster_mask], cmap='viridis', s=50, label='Clustered points')

# Plot outliers/noise
plt.scatter(X_scaled[noise_mask, 0], X_scaled[noise_mask, 1],
            c='red', s=50, marker='x', label='Outliers')

plt.title('DBSCAN Clustering on Moon-Shaped Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# 29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

# Import libraries
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Evaluate silhouette scores for k = 2 to 5
silhouette_scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    y_kmeans = kmeans.fit_predict(X)
    score = silhouette_score(X, y_kmeans)
    silhouette_scores.append(score)

# Plot silhouette scores as a bar chart
plt.figure(figsize=(8,6))
plt.bar(k_values, silhouette_scores, color='skyblue')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different K in K-Means')
plt.xticks(k_values)
plt.show()
```

```python
# 30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.

from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage

#  Load the Iris dataset
iris = load_iris()
X = iris.data

# Perform hierarchical clustering using average linkage
linked = linkage(X, method='average')

# Plot the dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linked,
           orientation='top',
           distance_sort='ascending',
           show_leaf_counts=True,
           labels=iris.target)
plt.title('Hierarchical Clustering Dendrogram (Average Linkage)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
# 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=2.0, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title('K-Means Clustering with Decision Boundaries')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# 44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

from sklearn.datasets import load_iris
from sklearn.metrics import silhouette_samples, silhouette_score

iris = load_iris()
X = iris.data

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

sil_scores = silhouette_samples(X, labels)
avg_score = silhouette_score(X, labels)
print(f"Average Silhouette Score: {avg_score:.3f}")

plt.figure(figsize=(8,6))
y_ticks = np.arange(len(X))
plt.barh(y_ticks, sil_scores, color='skyblue')
plt.axvline(avg_score, color='red', linestyle='--', label=f'Average Score: {avg_score:.2f}')
plt.title('Silhouette Scores for Each Sample (Iris Dataset)')
plt.xlabel('Silhouette Coefficient')
plt.ylabel('Sample Index')
plt.legend()
plt.show()
```

```python
# 47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

from sklearn.datasets import make_blobs

X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=1.5, random_state=42)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dbscan = DBSCAN(eps=0.8, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = np.sum(labels == -1)

print("Number of clusters (excluding noise):", n_clusters)
print("Number of noise points:", n_noise)
```

```python
# 48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

from sklearn.datasets import load_digits
from sklearn.manifold import TSNE

digits = load_digits()
X = digits.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

tsne = TSNE(n_components=2, random_state=42, perplexity=30, learning_rate=200)
X_embedded = tsne.fit_transform(X_scaled)

agg = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg.fit_predict(X_embedded)

plt.figure(figsize=(8,6))
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=labels, cmap='tab10', s=40)
plt.title('Agglomerative Clustering on Digits Dataset (t-SNE 2D)')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```
